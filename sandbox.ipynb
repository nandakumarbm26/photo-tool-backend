```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img=cv2.imread('input2.jpg',cv2.IMREAD_UNCHANGED)

dst=cv2.fastNlMeansDenoisingColored(img,None,5,10,7,21)
kernel = np.array([[0, -1, 0],
                   [-1, 5,-1],
                   [0, -1, 0]])
sharp = cv2.filter2D(img,-1,kernel)
sharpdst = cv2.filter2D(dst,-1,kernel)
plt.subplot(221),plt.imshow(img)
plt.subplot(222),plt.imshow(dst)
plt.subplot(223),plt.imshow(sharp)
plt.subplot(224),plt.imshow(sharpdst)
plt.show()
cv2.imwrite('outdst.jpg',sharpdst)
```
